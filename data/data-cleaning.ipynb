```python
import pandas as pd
import numpy as np
```

```python
student_port = pd.read_csv('student-por.csv')
```

```python
student_port.head()
```

Information about the data:

- School: GP: Gabriel Pereira or MS: Mousinho da Silveira
- sex: F-female M- male
- age: from 15 to 22
- address: U- urban R- rural
- famsize: family size- LE3: less or equal to 3 GT3- greater than 3
- Pstatus: parent cohabitation status: T - living together A - appart
- Medu: mother's education level ( 0-non, 1- primary education, 2 - 5th to 9th Grade, 3 - secondary education, 4 - higher education)
- Fadu: father's education level
- Mjob: mother's job (teacher, health related, civil services, at_home, other)
- Fjob: father's job
- reason: reason to choose this school (close to home, school reputation, course preference, other)
- guardian - student's guardian (mother, father, others)
- traveltime - from home to school (hours)
- studytime - weekly study time (hours)
- failures - number of past class failures (numeric: n if 1<=n<3, else 4)
- schoolsup - extra educational support (binary: yes or no)
- famsup - family educational support (binary: yes or no)
- paid - extra paid classes within the course subject (Math or Portuguese) (binary: yes or no)
- activities - extra-curricular activities (binary: yes or no)
- nursery - attended nursery school (binary: yes or no)
- higher - wants to take higher education (binary: yes or no)
- internet - Internet access at home (binary: yes or no)
- romantic - with a romantic relationship (binary: yes or no)
- famrel - quality of family relationships (numeric: from 1 - very bad to 5 - excellent)
- freetime - free time after school (numeric: from 1 - very low to 5 - very high)
- goout - going out with friends (numeric: from 1 - very low to 5 - very high)
- Dalc - workday alcohol consumption (numeric: from 1 - very low to 5 - very high)
- Walc - weekend alcohol consumption (numeric: from 1 - very low to 5 - very high)
- health - current health status (numeric: from 1 - very bad to 5 - very good)
- absences - number of school absences (numeric: from 0 to 93)

- G1 - first period grade (numeric: from 0 to 20)
- G2 - second period grade (numeric: from 0 to 20)
- G3 - final grade (numeric: from 0 to 20, output target)


#### 1. Change column names


```python
student_port = student_port.rename(columns = {'famsize' : 'family_size', 'Pstatus': 'parent_status', 'Medu': 'mothers_education', 'Fedu':'fathers_education', 'Mjob': 'mothers_job', 'Fjob': 'fathers_job', 'schoolsup': 'extra_support', 'famsup': 'family_support', 'paid': 'extra_classes', 'higher': 'higher_education', 'famrel': 'family_relation', 'goout': 'out_with_friends', 'Dalc': 'daily_alcohol_cons', 'Walc': 'weekend_alcohol_cons', 'G1': 'first_grade', 'G2': 'second_grade', 'G3': 'final_grade'})
```

```python
student_port.head()
```

```python
# check for isna values

student_port.isna().sum()
```

```python
# replace school names

student_port["school"] = student_port["school"].replace({'GP': 'Gabriel Pereira', 'MS': 'Mousinho da Silveira'})
```

```python
student_port.head()
```

```python
# How many students in each school participate?

student_port["school"].value_counts()
```

```python
# Change sex from initial to full description

student_port["sex"] = student_port["sex"].replace({'F': 'Female', 'M': 'Male'})
```

```python
student_port.head()
```

```python
# how many students of each sex participate? 

student_port["sex"].value_counts()
```

```python
# change the values for address from initial to full information

student_port["address"] = student_port["address"].replace({'U': 'Urban', 'R': 'Rural'})
```

```python
student_port.head()
```

```python
# How many students from each of the address types?

student_port["address"].value_counts()
```

```python
# change values from family_size column to something that describe it better.

student_port["family_size"] = student_port["family_size"].replace({'GT3': '4 or more', 'LE3': '3 or less'})
```

```python
student_port.head()
```

```python
# are the values equal?

student_port["family_size"].value_counts()
```

```python
# change the parent_status column information to something that better describe the data.

student_port["parent_status"] = student_port["parent_status"].replace({'A': 'Apart', 'T': 'Together'})
```

```python
student_port.head()
```

```python
student_port["parent_status"].value_counts()
```

```python
# how is the daily_alcohol_cons frequency?

student_port["daily_alcohol_cons"].value_counts()
```

```python
# and the weekend_alcohol_cons frequency?

student_port["weekend_alcohol_cons"].value_counts()
```

```python
# change the parent's education level from numeric to descriptive string

student_port["mothers_education"] = student_port["mothers_education"].replace({0: 'None', 1: 'primary', 2: '5th - 9th', 3:'secondary', 4: 'higher'})
student_port["fathers_education"] = student_port["fathers_education"].replace({0: 'None', 1: 'primary', 2: '5th - 9th', 3:'secondary', 4: 'higher'})
```

```python
# change other numerical data into descriptive string

student_port["family_relation"] = student_port["family_relation"].replace({1: 'very bad', 2: 'bad', 3:'normal', 4: 'good', 5: 'excellent'})
```

```python
student_port["freetime"] = student_port["freetime"].replace({1: 'very low', 2: 'low', 3:'normal', 4: 'high', 5: 'very high'})
```

```python
student_port["out_with_friends"] = student_port["out_with_friends"].replace({1: 'very low', 2: 'low', 3:'normal', 4: 'high', 5: 'very high'})
```

```python
student_port["daily_alcohol_cons"] = student_port["daily_alcohol_cons"].replace({1: 'very low', 2: 'low', 3:'normal', 4: 'high', 5: 'very high'})
```

```python
student_port["weekend_alcohol_cons"] = student_port["weekend_alcohol_cons"].replace({1: 'very low', 2: 'low', 3:'normal', 4: 'high', 5: 'very high'})
```

```python
student_port["health"] = student_port["health"].replace({1: 'very bad', 2: 'bad', 3:'normal', 4: 'good', 5: 'excellent'})
```
